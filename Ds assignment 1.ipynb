```python
#BOXPLOT-FINDING OUTLIERS
#we already have median,q1,q3,iqr
#calculating inner fences
f1=q1-((1.5)*(iqr))
f3=q3+((1.5)*(iqr))

```

```python
a1=0
a2=0
diff1=float('Inf')
diff2=float('Inf')
for i in np.nditer(data):
    if f1<0:
        a1=0
    elif (i-f1+1)<diff1 and (i-f1+1)>0:
        diff1=i-f1+1
        a1=i
for i in np.nditer(data):
    #global diff2= float('Inf')
    #global a2=0
    if (f3-i+1)<diff2 and (f3-i+1)>0:
        diff2=f3-i+1
        a2=i

```

```python
#calculating outer fences
F1=q1-(2*(1.5)*(iqr))
F3=q3+(2*(1.5)*(iqr))

```

```python
#finding mild and extreme outliers
for i in np.nditer(data):
    if (i<a1 and i>F1) or (i>a2 and i<F3):
        print("Mild Oulier: ",i)
    elif(i<F1) or (i>F3):
        print("Extreme Outlier ",i)
```
